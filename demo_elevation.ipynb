# Build a local Digital Elevation Model (DEM) from IGN BD ALTI to get elevation over an area of interest

```python
from dem import Dem
```

```python
folder = "BDALTIV2_MNT_75M_ASC_LAMB93_IGN69_FRANCE"  # link to your BD ALTI ASCII folder
center = (43.60426, 1.44367)  # Toulouse, France
radius = 5100  # meters (multiple of 75 m)
```

#### <br> Find and plot the BD ALTI tile covering the area of interest

```python
d = Dem(folder, center, radius)
d.plot2D_tile()
```

#### <br> Plot the DEM of the area of interest (2D view)

```python
d.plot2D(color_range=(0.08, 0.8))
```

#### <br> Plot the DEM of the area of interest (3D view)

```python
d.plot3D(color_range=(0.08, 0.8))
```

#### <br> Get elevation of points within this area of interest

```python
print(d.get_elevation(43.596811, 1.408147))
print(d.get_elevation(43.565097, 1.445938))
```
